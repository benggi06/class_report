```python
for i in range(1,101):
    if i % 3 == 0:
        print('Fizz', end='')
        if i % 5== 0:
            print('Buzz')
        else:
            print()
    elif i % 5 == 0:
        print('Buzz')
            
    else:
        print(i)
```

```python
for i in range(1,101):
    if i % 3==0 and i % 5 ==0:
        print('FB')
    elif i % 3 == 0:
        print('F')
    elif i % 5 == 0:
        print('B')
    else:
        print(i)
```
